```python
def map_labels(int_label): return {0 : "Not Spam", 1:"Spam"}[int_label]

def test_model(classifier, test_data):

  """
    Parameters:
      classifier : classification model
      test_data : testing data

    return:
      predicted values
  """
  try:
    predicted_values = classifier.predict(test_data)
    return predicted_values
  except Exception as e:
    print(e)
    return None


def classify(classifier, test_data):
  try:
    predicted_values = classifier.predict([test_data])
    return predicted_values[0]
  except Exception as e:
    print(e)
    return None



def Evaluate_classification_model(actual_values, predicted_values):
  """
  Parameters:
    actual_values : True value for data
    predicted_values : Predicted values by the model

  return :
      return dictionary with confusion matrix,
                             accuracy, precision, recall, F1-score
  """
  try :
    return {
        "cinfusion_matrix" : confusion_matrix(predicted_values, actual_values),
        "accuracy" : accuracy_score(predicted_values, actual_values),
        "precision" : precision_score(predicted_values, actual_values),
        "recall" : recall_score(predicted_values, actual_values),
        "F1" : f1_score(predicted_values, actual_values)
    }

  except Exception as e:
    print(e)
    return None


def visualize_cinfusion_matrix(confusion_matrix, labels):
  """
    viszualize confusion matrix as a heat-map
    Parameters:
      confusion_matrix : array that represent confusion matrix
      labels : class labels, or models outoputs
    return :


  """

  try:
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,
                                  display_labels=labels)
    disp.plot()
    plt.show()

  except Exception as e:
    print(e)
    return None


# Save ML model
def save_model(model, path):
  try:
    pickle.dump(model, open(path, 'wb'))
    print("model saved successfully")
  except Exception as e:
    print(e)
    return None

# Load ML model
def load_model(path):
  try:
    loaded_model = pickle.load(open(path, 'rb'))
    return loaded_model
  except Exception as e:
    print(e)
    return None
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay # Importing necessary functions from sklearn.metrics
import pickle
# 1- Build Multi-Nimial Naive bayes model
MNBClassifier = MultinomialNB()

# 2- Fit data into a model
MNBClassifier.fit(X_train, y_train) # training

# 3- Predict testing data
y_pred = test_model(classifier=MNBClassifier , test_data=X_test)

# 4- Evaluate model
dict_evaluation_info = Evaluate_classification_model(actual_values=y_test , predicted_values=y_pred)
print("\n", pd.Series(dict_evaluation_info), "\n")

# 5- Visualize confusion Matrix
visualize_cinfusion_matrix(confusion_matrix=dict_evaluation_info["cinfusion_matrix"],
                           labels=MNBClassifier.classes_)

path_MNBClassifier_model = "./MNBClassifier.pkl"
save_model(model=MNBClassifier, path = path_MNBClassifier_model)
```

```python
def prepare_text_for_prediction(dict_emails, text):
  """
  This function used to convert input text into a formate that used in prediction
  by represent each word with it's occurrencies
  parameters :
    dict_emails : initiated dictionary where the keys represent the words in the model
  return :
    return list of values that represent email info

  """
  try:
    for word in text.split() :
      if word in dict_emails.keys():
        dict_emails[word] +=1

    return np.array(list(dict_emails.values()))

  except Exception as e:
    print(e)
    return None
```
